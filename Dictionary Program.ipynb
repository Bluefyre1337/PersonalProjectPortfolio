```python
def writeFileAndTerminal(pString, pFile, pFileNewline):
  print(pString)
  pFile.write(pString+pFileNewline)
```

```python
def genPrefix(pNumOfSpaces):
  stringArray = []
  for i in range(pNumOfSpaces):
    stringArray.append(" ")
  stringArray.append("● ")
  return ("".join(stringArray))
```

```python
def getMeaning(pWord):
  returnMeaning = dict.meaning(pWord)
  if returnMeaning is None:
    undefinedList.append(pWord)
  return returnMeaning
```

```python
def printMeaning(pMeaning):
  if pMeaning is None:
    print(f'####DEFINITION NOT FOUND')
  else:
    for wordTypeKey in pMeaning:
      print(f'{wordTypeKey.upper()}')
      definitionGroup = (pMeaning.get(wordTypeKey))
      for definition in definitionGroup:
        print(f'{genPrefix(2)}{definition.capitalize()}')
```
